```python
# Inhertiance : used to inherit from the main class. 

class Employee:
    def work(self):
        print("Working")
        
class Developer():
    def develop(self):
        print("Coding")
        
class Tester():
    def test(self):
        print("Testing")
        
obj1 = Tester()
obj1.test()
obj2 = Developer()
obj2.develop()
```

```python
#Bubble Sort . 

arr = [10 , 8, 6, 4, 2, 1]
n = len(arr)
for i in range(n):
    for j in range(0 ,n-i-1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print(arr)

```
